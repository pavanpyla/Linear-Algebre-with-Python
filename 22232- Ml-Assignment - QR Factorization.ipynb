## Calculating the Least Square Solution using  QR  Factorization 

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import qr as QR                                        
from numpy.linalg import inv
from mpl_toolkits.mplot3d import Axes3D
```

Enter any matrix consisting of n-1 predictoes (features) and 1 response varianble in last column.
Then below code will give the Least squares estimation using QR factorization 

```python
m=int(input("Enter the no of rows : "))
n=int(input("Enter the no of columns : "))
A=np.zeros(n*m)
A=A.reshape(m,n)                       # Enter the any matrix with n-1 predicto variables and nth column as Response variable

for i in range(0,m):
    for j in range(0,n):
        A[i,j]=float(input(f"enter the element A[{i}][{j}] : "))
```

```python
X=A[ : , :-1]
Y=A[ : , -1]
```

```python

print(X,Y)                   # Printing the X and Y 
```

```python
Q , R = QR(X)           # Calculating the Q and R
```

```python
print(Q,R)               # Printing the Q and R Matrices
```

```python
print(np.dot(Q,R))     # checking Q R factorization
```

```python
iR=inv(R)                   # calculating inverse of R and Transpose of Q
iQ=(Q.T)
```

```python
inter = iQ.dot(Y)                   # calculating inter mediate results betweeen  Q.Y
inter
```

```python
w = iR.dot(inter)                # calculating 
w                               
```

### Å· = w . x 

```python
yhat = X.dot(w)            # dot product between calculated w and x

yhat
```

## Plotting the Results 2D  and 3D

```python
if(n<=2):
    plt.figure(figsize=(8, 5))
    plt.xlabel("X - axis",size=20)
    plt.ylabel("y - axis",size=20)
    plt.scatter(X,Y,label=' Data points ',color='red')
    plt.plot(X,yhat,color='green',label= "Least Square Estimation")
    plt.legend()
elif(n==3):
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, projection='3d')
    
    ax.scatter3D(X[:,0] ,X[:,1] ,yhat, marker='o', s=40,color='red')
    ax.plot(X[:,0] ,X[:,1] ,Y)

    plt.show()
else:
    print("Cannot plot more than 3 dimension")
```
